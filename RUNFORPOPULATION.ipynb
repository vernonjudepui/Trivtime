```python
%%sql
DROP DATABASE IF EXISTS TRIVTIME;
CREATE DATABASE TRIVTIME;
USE TRIVTIME;
CREATE TABLE LEVEL(
    LVL INTEGER NOT NULL,
    EXPTOLEVEL INTEGER NOT NULL,
    PRIMARY KEY(LVL)
);
CREATE TABLE USER(
    USERID INTEGER NOT NULL,
    
    
    UserName VARCHAR(30),
    Elo	INTEGER NOT NULL,
    Points INTEGER NOT NULL,
    LVL INTEGER NOT NULL,
    ExpToNextLeveL INTEGER NOT NULL,
    CurrExp INTEGER NOT NULL,
    Password VARCHAR(200),
    EMAIL VARCHAR(60),
    PRIMARY KEY(USERID),
    UNIQUE ( USERNAME ),
    FOREIGN KEY (LVL) REFERENCES LEVEL(LVL)  ON DELETE CASCADE
);
CREATE TABLE GAME(
    GAMEID INTEGER NOT NULL,
    QTYPE INTEGER NOT NULL,
    NumDIFF INTEGER NOT NULL,
    NUMQ INTEGER NOT NULL,
    PRIMARY KEY(GAMEID)
    
    
    
    
);

CREATE TABLE ROOM(
    RoomID		INTEGER NOT NULL AUTO_INCREMENT,
    RoomKey		VARCHAR(20),
    HOST INTEGER,
    MaxPlayers		INTEGER NOT NULL,
    NumPlayers 	INTEGER NOT NULL,
    PRIMARY KEY(ROOMID),
    FOreign KEY (HOST) REFERENCES USER(USERID) ON DELETE CASCADE
    
    
);

CREATE TABLE BADGES(
    BID	INTEGER NOT NULL,
    BadgeTitle	VARCHAR(20) NOT NULL,
    BadgeDes VARCHAR(100) NOT NULL,
    PRIMARY KEY(BID)
    
);
CREATE TABLE ACHIEVEMENTS(
    AID	INTEGER NOT NULL,
 
    
    AchieveTitle VARCHAR(30) NOT NULL,
       AchieveDes	VARCHAR(100) NOT NULL,
    NumPlayersAwarded INTEGER NOT NULL,
    
     PRIMARY KEY(AID)
    
);
CREATE TABLE QUESTION(
    QID INTEGER NOT NULL,
    Difficulty	INTEGER NOT NULL,
    Category	VARCHAR(30) NOT NULL,
    QuestionText VARCHAR(1024),	
    Qtype INTEGER NOT NULL,
    PRIMARY KEY(QID)
    
);
CREATE TABLE PICTURE(
     QID INTEGER NOT NULL,
     IMAGESOURCE varbinary(32786),
     ANS VARCHAR(64),
     PRIMARY KEY(QID),
     FOREIGN KEY (QID) REFERENCES QUESTION(QID) ON DELETE CASCADE
    
);
CREATE TABLE UCSTYLE(
     QID INTEGER NOT NULL,
     
     ANS VARCHAR(64),
     PRIMARY KEY(QID),
     FOREIGN KEY (QID) REFERENCES QUESTION(QID) ON DELETE CASCADE
    
);
CREATE TABLE MCQ(
     QID INTEGER NOT NULL,
     
     ANSCHOICE VARCHAR(512),
    CHOICES INTEGER NOT NULL,
     PRIMARY KEY(QID),
     FOREIGN KEY (QID) REFERENCES QUESTION(QID) ON DELETE CASCADE
    
);
CREATE TABLE TEXT(
    
    
      QID INTEGER NOT NULL,
     
     ANS VARCHAR(64),
     PRIMARY KEY(QID),
     FOREIGN KEY (QID) REFERENCES QUESTION(QID) ON DELETE CASCADE
    
);

CREATE TABLE GIVES(
          QID INTEGER NOT NULL,
     
    GAMEID INTEGER NOT NULL,
    NUMBER INTEGER NOT NULL,
    
     PRIMARY KEY(QID,GAMEID),
    FOREIGN KEY(GAMEID) REFERENCES GAME(GAMEID) ON DELETE CASCADE,
    
     FOREIGN KEY (QID) REFERENCES QUESTION(QID) ON DELETE CASCADE
    
    
);
CREATE TABLE JOINEDROOM(
    JOINEDID INTEGER NOT NULL AUTO_INCREMENT,
    ROOMID INTEGER NOT NULL,
    USERID INTEGER NOT NULL,
    PRIMARY KEY(JOINEDID),
      FOREIGN KEY(ROOMID) REFERENCES ROOM(ROOMID) ON DELETE CASCADE,
    FOREIGN KEY(USERID) REFERENCES USER(USERID) ON DELETE CASCADE



);
CREATE TABLE PLAYS(
    GameID	 INTEGER NOT NULL,
    JOINEDID INTEGER NOT NULL,
    
    StartTime	 DATETIME NOT NULL,
    EndTime  DATETIME NOT NULL,
     PRIMARY KEY(JOINEDID,GAMEID),
     FOREIGN KEY(GAMEID) REFERENCES GAME(GAMEID) ON DELETE CASCADE,
         FOREIGN KEY(JOINEDID) REFERENCES JOINEDROOM(JOINEDID) ON DELETE CASCADE
);
CREATE TABLE SCOREIN(
    GameID	 INTEGER NOT NULL,

    UserID	 INTEGER NOT NULL,
    
    Score INTEGER NOT NULL,
     PRIMARY KEY(GAMEID,USERID),
     FOREIGN KEY(GAMEID) REFERENCES GAME(GAMEID) ON DELETE CASCADE,
   
    FOREIGN KEY(USERID) REFERENCES USER(USERID) ON DELETE CASCADE
    
);
CREATE TABLE EARNS(
    UserID	 INTEGER NOT NULL,
    BID INTEGER NOT NULL,
     PRIMARY KEY(UserID,BID ),
        FOREIGN KEY(USERID) REFERENCES USER(USERID) ON DELETE CASCADE,
    FOREIGN KEY (BID) REFERENCES BADGES(BID) ON DELETE CASCADE
    
    
    
);
CREATE TABLE ACHIEVES(
      UserID	 INTEGER NOT NULL,
    AID INTEGER NOT NULL,
     PRIMARY KEY(UserID,AID ),
        FOREIGN KEY(USERID) REFERENCES USER(USERID) ON DELETE CASCADE,
    FOREIGN KEY(AID) REFERENCES ACHIEVEMENTS(AID) ON DELETE CASCADE
    
    
);
CREATE TABLE ANSWERS(
     
      GAMEID INTEGER NOT NULL,
      QID INTEGER NOT NULL,
     UserID	 INTEGER NOT NULL,
      TIMETAKEN time NOT NULL,
    
     PRIMARY KEY(UserID,GAMEID,QID ),
        FOREIGN KEY(USERID) REFERENCES USER(USERID) ON DELETE CASCADE,
      FOREIGN KEY(GAMEID) REFERENCES GAME(GAMEID) ON DELETE CASCADE,
         FOREIGN KEY (QID) REFERENCES QUESTION(QID) ON DELETE CASCADE
    
);
CREATE TABLE HOSTS(
    ROOMID INTEGER NOT NULL,
    USERID INTEGER NOT NULL,
    PRIMARY KEY(UserID,ROOMID ),
    FOREIGN KEY(USERID) REFERENCES USER(USERID) ON DELETE CASCADE,
    FOREIGN KEY(ROOMID) REFERENCES ROOM(ROOMID) ON DELETE CASCADE
    
    
);
CREATE TABLE CATEGORIES(
    GAMEID INTEGER NOT NULL,
    CATEGORY VARCHAR(30) NOT NULL,
    PRIMARY KEY(GAMEID,CATEGORY),
    FOREIGN KEY(GAMEID) REFERENCES GAME(GAMEID) ON DELETE CASCADE
    
);
show tables;
```
